```python
import string
import requests
import random

result = requests.get('https://raw.githubusercontent.com/dwyl/english-words/master/words.txt')

text_list = text_list = [word.upper() for word in result.text.splitlines() if word.isalpha()]

hangman={0:'''
        ____________
         |''',
        1:'''
        ____________
         |
         O''',
        2:'''
        ____________
         |
         O
        /''',
        3:'''
        ____________
         |
         O
        / \\''',
        4:'''
        ____________
         |
         O
        / \\
         |''',
        5:'''
        ____________
         |
         O
        / \\
         |
        /''',
        6:'''
        ____________
         |
         O
        / \\
         |
        / \\ '''}

while True:
    try:
        level = int(input(f'Choose a level by typing 1 for Easy (Get 2 of all letters when only 2 mistakes left),\n 2 for Medium (Get 1 of all letters when only 2 mistakes left),\n or 3 for Hard (No Help) (insert only integers):'))
        if level == 1:
            level=1
        elif level == 2:
            level = 2
        elif level == 3:
            level = 3
        else:
            print('Please chose either 1, 2, or 3.')
            continue
        break
    except ValueError:
        print('Invalid input. Please enter only integers.')

word = random.choice(text_list)
hidden_word = '-'*len(word)
print(f'You chose level: {level} and your hidden word is: {hidden_word}')


mistake_count = 6
print('==============================================================')
guessed_letters = ""
help_letters_count = 0

def hangman_function():
    global guessed_letters, hidden_word, mistake_count
    picked_letter = input("Please insert a single letter: ", ).capitalize()
    if picked_letter.isalpha() and len(picked_letter)==1:
        if picked_letter in guessed_letters:
            print("You already guessed this letter.")
            print('==============================================================')

        else:
            guessed_letters += picked_letter
            if picked_letter in word:
                print(f"Correct: The hidden word contains the letter {picked_letter}")
                index_count = 0
                for letter in word:
                    if letter == picked_letter:
                        hidden_word = hidden_word[:index_count] + picked_letter + hidden_word[index_count+1:]
                        #letter_count += 1
                    index_count += 1
            else:
                mistake_count -= 1
                print(f"Incorrect: The hidden word does not contain the letter {picked_letter}. Number of mistakes left: {mistake_count}.")
                print(hangman[6 - mistake_count])
        print(hidden_word)
        print('==============================================================')
    else:
        print("Please use only one letter from the alphabet.")
        print('==============================================================')

def help_function(i):
    global guessed_letters, help_letters_count, help_letter, hidden_word
    help_letters = [letter for letter in word if letter not in guessed_letters]
    if len(help_letters) > i:
        letter_chosen = random.choice(help_letters)
        guessed_letters += letter_chosen

        index_count = 0
        for letter in word:
            if letter == letter_chosen:
                hidden_word = hidden_word[:index_count] + letter_chosen + hidden_word[index_count+1:]
            index_count += 1
        help_letters_count += 1
        print(f'The hint you are receiving is: {letter_chosen}.')
        print(hidden_word)
        print('==============================================================')
    else:
        print('No more hints available')


if level == 3:
    while mistake_count > 0 and hidden_word.count('-') != 0:
        hangman_function()
elif level == 2:
    while mistake_count > 2 and hidden_word.count('-') != 0:
        hangman_function()
    if mistake_count == 2 and hidden_word.count('-') != 0:
        help_function(1)
    else:
        print("There is only one letter you are missing. Therefore, there are not hints available for you.")
    while mistake_count > 0 and hidden_word.count('-') != 0:
        hangman_function()
elif level == 1:
    while mistake_count > 2 and hidden_word.count('-') != 0:
        hangman_function()
    if mistake_count == 2 and hidden_word.count('-') != 0:
        help_function(2)
        help_function(1)
    else:
        print("There is only one letter you are missing. Therefore, there are not hints available for you.")
    while mistake_count > 0 and hidden_word.count('-') != 0:
        hangman_function()


if mistake_count == 0:
  print("HANGED!!")
  print(f'The word would have been: {word}')
else:
  print('CONGRATULATIONS!!!')
```
